### pivot_table

```python
# 각 state마다 price의 평균

pd.pivot_table(order, index='state',values='price')
```

```python
pd.pivot_table(order, index='address',values='price')
```

```python
# aggfunc= ?     2개 이상 보고 싶으면 list로 만들면 됨

pd.pivot_table(order, index='state', values='price', aggfunc=['sum','mean'])
```

### group by

```python
order.price.mean()
```

```python
order.groupby('state')['price'].mean()
```

```python
pd.pivot_table(order,index='state',values='price')
```

```python
order.groupby(['state','date'])['price'].mean()
```

pivot_table vs group by   차이점이 뭐야????????  
  
  + 피봇테이블은 주로 통계값을 보고 싶을 때 진행 (출력값이 예쁘게 나옴)  
  + 그룹바이는 묶어서 이제 다음 진행을 위한 작업으로 진행

```python
%timeit order.groupby(['state','date'])['price'].mean()
```

```python
import numpy as np
%timeit pd.pivot_table(order,index=['state','date'],values='price',aggfunc=np.mean)
```

앞선 두 결과를 보시다시피 pivot_table보다 groupby가 실행속도가 더 빠르다.  
그렇기 때문에 출력값을 바로 요하지 않을 때에는 groupby를 사용하는 것이 더 좋다.

### apply

```python
# order["price"]만 apply

def is_vip(price):
    if price>=500:
        return 'VIP'
    else:
        return 'Non-VIP'

order['price'].apply(is_vip)
```

```python
# order 전체에 apply

def is_vip(row):
    if row.state=='confirmed' and row.price>=500:
        return "VIP"
    else:
        return "Non-VIP"

order.apply(is_vip,axis="columns")
```

### 여러 개의 데이터를 하나로 합치기

```python
url = "https://bit.ly/dsa-02-order-201701"

order201701=pd.read_csv(url,index_col='id')
order201701
```

```python
url = "https://bit.ly/dsa-02-order-201702"

order201702=pd.read_csv(url,index_col='id')
order201702
```

```python
### concat 데이터를 조각조각 나뉘어있을때 사용하면 good!

order=pd.concat([order201701,order201702])
order
```

```python
url="https://bit.ly/dsa-02-order-how"
how=pd.read_csv(url,index_col="id")
how
```

```python
# concat axis의 default값은 위아래(index) //  axis="columns" 는 좌우로(columns)

pd.concat([order,how],axis="columns")
```

concat의 기준은 index! 따라서, concat을 할 때 합치는 데이터셋의 index를 설정하는 것이 굉장히 중요하다!!!

```python
url="https://bit.ly/dsa-02-user"
user=pd.read_csv(url,index_col='id')
user
```

```python
url="https://bit.ly/dsa-02-product"
product=pd.read_csv(url,index_col='id')
product
```

```python
# order 랑 user 합치기 

pd.merge(order,user,left_on='user_id',right_on='id')
```

left_on, right_on을 써서 데이터를 합칠 때 동일한 기준을 적용. 

```python
# 2개의 데이터를 합치지만 기준 column명이 동일할 때

user['user_id']=user.index
pd.merge(order,user,on='user_id')
```

```python
# user_id를 제외하고 동일한 key가 없기 때문에 지정을 안해도 자동으로 user_id를 기준으로 합침

pd.merge(order,user)
```

```python
pd.merge(order,product,left_on='product_id',right_on='id')
```

기준을 잡아서 order랑 product를 합쳤지만 price라는 동일한 column명을 갖고 있었기 때문에  
price_x, price_y 가 생기게 된다. 이럴 경우, rename을 하거나 둘 중 하나를 살리는 방법 등을 취할 수 있다.

```python
# how가 default는 how="inner"  inner는 교집합 즉, 왼쪽 오른쪽 둘 다 있어야 함.

pd.merge(order,user,on="user_id", how="inner")
```

```python
# how="outer"  outer는 합집합 즉, 왼쪽 오른쪽 중 한쪽에만 있어도 살림.

pd.merge(order,user,on="user_id", how="outer")
```

```python
# how="left"  left 왼쪽을 살림.

pd.merge(order,user,on="user_id", how="left")
```

```python
# how="right"  right 오른쪽을 살림.

pd.merge(order,user,on="user_id", how="right")
```

+ TIP. merge할 때  
메인 데이터를 left에 놓고 무조건 left merge하는 것을 추천

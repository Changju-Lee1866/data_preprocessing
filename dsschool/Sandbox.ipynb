```python
import sqlite3
import pandas as pd
```

# Load Data

```python
connection=sqlite3.connect("e-commerce.sqlite")
connection
```

sqlite는 오류를 찾아내기 어려움. 만약에 위 구문에서 "e-commerce.db"라고 실행시켰다면  
오류는 없겠지만 e-commerce.db라는 파일은 없으므로 자동으로 만들어 낸다.  
그 후, 아래에서 실행하는 문구에서 오류가 되게 되는 것이다.

```python
query= "SELECT * FROM 'orders'"

order= pd.read_sql(query,connection, index_col='id')

order
```

```python
query= "SELECT id,date,price FROM 'orders'"

order= pd.read_sql(query,connection, index_col='id')

order
```

```python
query= "SELECT * FROM 'orders' limit 3"

order= pd.read_sql(query,connection, index_col='id')

order
```

```python
query= "SELECT * FROM 'orders' ORDER BY date ASC"

order= pd.read_sql(query,connection, index_col='id')

order
```

```python
query= "SELECT * FROM 'orders' ORDER BY date DESC"

order= pd.read_sql(query,connection, index_col='id')

order
```

```python
query= """
    SELECT * 
        FROM 'orders' 
        ORDER BY date ASC
        LIMIT 3
    
"""
order= pd.read_sql(query,connection, index_col='id')

order
```

```python
query= """
    SELECT *
        FROM 'orders'
        WHERE state <> 'confirmed'
"""
# <> --> != 이랑 같다. SQL에서는 <>을 권장.

order= pd.read_sql(query,connection, index_col='id')

order
```

```python
query= """
    SELECT *
        FROM 'orders'
        WHERE price BETWEEN 400 AND 900 
"""
# <> --> != 이랑 같다. SQL에서는 <>을 권장.

order= pd.read_sql(query,connection, index_col='id')

order
```

```python
query= """
    SELECT *
        FROM 'orders'
        WHERE address IN ('Incheon','Busan') 
"""
# <> --> != 이랑 같다. SQL에서는 <>을 권장.

order= pd.read_sql(query,connection, index_col='id')

order
```

```python
query= """
    SELECT *
        FROM 'orders'
        WHERE address LIKE 'n%' 
"""
# %n% n을 포함하는 // %n n으로 끝나는 // n% n으로 시작하는

order= pd.read_sql(query,connection, index_col='id')

order
```
